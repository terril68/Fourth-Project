### scraping this page: https://en.wikipedia.org/wiki/List_of_corporations_in_Pittsburgh  for a list of Fortune 500 companies in Pittsburgh

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_corporations_in_Pittsburgh')
doc = BeautifulSoup(response.text)
doc
```

```python
doc.find_all("li")
```

```python
for lists in doc.find_all("li"):
    print (lists.text.strip())
    print (lists.find("a")['href'])
    print ('----')
```
