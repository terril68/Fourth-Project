```python
response = requests.get('https://www.bls.gov/emp/tables/occupational-projections-and-characteristics.htm')
doc = BeautifulSoup(response.text)
doc
```

```python
tables = pd.read_html(response.text, flavor='html5lib')

tables[0]
```

## Cleaning the table

```python
df=tables[0]
```

```python
df.to_csv("bls.csv", index=False)
```

```python
df = pd.read_csv("bls.csv",na_values="—")
```

```python
df.head()
```

```python
df.dtypes
```

## changing the median annual wage, 2021(1) column to integers

```python
df['Median annual wage, 2021(1)']=df['Median annual wage, 2021(1)'].str.strip('$')
```

```python
df['Median annual wage, 2021(1)']=df['Median annual wage, 2021(1)'].str.replace(',','')
```

```python
df['Median annual wage, 2021(1)']=df['Median annual wage, 2021(1)'].replace('>=$208000','208000')
```

```python
df[df['Median annual wage, 2021(1)'].isnull()]
```

```python
df = df.dropna(subset=['Median annual wage, 2021(1)'])
```

```python
df['Median annual wage, 2021(1)'] = df['Median annual wage, 2021(1)'].astype(int)
```

```python
df.dtypes
```

## filtering the column to only display those occupations with salary above $50000, since that's the national average for college graduates, according to multiple sources

```python
df[df['Median annual wage, 2021(1)']>=50000]
```

```python
## save it into a new dataframe
df2=df[df['Median annual wage, 2021(1)']>=50000]
```

```python
df2
```

## filtering the Work experience in a related occupation column to show only those occupations requiring no prior experience

```python
df2[df2['Work experience in a related occupation']=='None']
```

```python
## saving it into a new dataframe
df3=df2[df2['Work experience in a related occupation']=='None']
```

```python
df3
```

## filtering the Typical education needed for entry column to only show those requiring bachelor degree

```python
df3[df3['Typical education needed for entry']=="Bachelor's degree"]
```

```python
## saving it into a new dataframe
df4=df3[df3['Typical education needed for entry']=="Bachelor's degree"]
```

## order the table by the fastest growing occupation

```python
df4
```

```python
df4.dtypes
```

### But first, the data type of the column 'Percent employment change, 2020–30' needs to be converted to float

```python
df4["Percent employment change, 2020–30"] = pd.to_numeric(df4["Percent employment change, 2020–30"])
```

```python
df4.dtypes
```

```python
df4.sort_values(by='Percent employment change, 2020–30',ascending=False).head(10)
```
